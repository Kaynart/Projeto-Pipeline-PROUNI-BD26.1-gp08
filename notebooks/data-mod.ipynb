# Definição das novas colunas

Os dados originais não possuem atributos suficientes que permitam a criação de uma dimensão temporal, logo serão adicionadas duas colunas: "DT_CONCESSAO_BOLSA" e "SEMESTRE_CONCESSAO_BOLSA", das quais será derivada a dimensão temporal necessária. Essa imputação ocorrerá de forma aleatória com base nas datas reais do Prouni de cada ano. 

Os seguintes valores serão adicionados em beneficiários aleatórios:

* Para o banco de dados Prouni-2017:

    1° semestre: 6 de fevereiro e 20 de fevereiro;

    2° semestre: 12 de junho e 26 de junho.

* Para o banco de dados Prouni-2018:

    1° semestre: 14 de fevereiro e 2 de março; 

    2° semestre: 2 de julho e 16 de julho.

* Para o banco de dados Prouni-2019:

    1° semestre: 6 de fevereiro e 20 de fevereiro;

    2° semestre: 18 de junho e 2 de julho;

```python
import pandas as pd
import numpy as np
```

```python
# Dicionário com os caminhos dos arquivos do ProUni
arquivos = {
    '2017': 'data_original/pda-prouni-2017-original.csv',
    '2018': 'data_original/pda-prouni-2018-original.csv',
    '2019': 'data_original/pda-prouni-2019-original.csv'
}

# Criação do dataframe para cada ano
df_2017 = pd.read_csv(arquivos['2017'], sep=';', 
            encoding='utf-8-sig',
            on_bad_lines='skip')

df_2018 = pd.read_csv(arquivos['2018'], sep=';', 
            encoding='utf-8-sig',
            on_bad_lines='skip')

df_2019 = pd.read_csv(arquivos['2019'], sep=';', 
            encoding='utf-8-sig',
            on_bad_lines='skip')
```

### Modificação dataset Prouni-2017 

```python
# Datas possíveis (padrão dd/mm/yyyy)
datas_1_semestre_2017 = ['06-02-2017', '20-02-2017']
datas_2_semestre_2017 = ['12-06-2017', '26-06-2017']

# Semestres possíveis incluídos de forma aleatória
df_2017['SEMESTRE_CONCESSAO_BOLSA'] = ""
df_2017['DT_CONCESSAO_BOLSA'] = ""

# Não adiciona novas informações em colunas nulas
linhas_validas = df_2017['ANO_CONCESSAO_BOLSA'].notna()
total_linhas_validas = linhas_validas.sum()

df_2017.loc[linhas_validas, 'SEMESTRE_CONCESSAO_BOLSA'] = np.random.choice(
    ['1º Semestre', '2º Semestre'], 
    size=total_linhas_validas
)

condicoes_17 = [
    df_2017['SEMESTRE_CONCESSAO_BOLSA'] == '1º Semestre',
    df_2017['SEMESTRE_CONCESSAO_BOLSA'] == '2º Semestre'
]

# Escolha das datas conforme semestre
escolhas_17 = [
    np.random.choice(datas_1_semestre_2017, size=len(df_2017)),
    np.random.choice(datas_2_semestre_2017, size=len(df_2017))
]

# Criação da coluna de data
df_2017['DT_CONCESSAO_BOLSA'] = np.select(condicoes_17, escolhas_17, default="")
df_2017.loc[~linhas_validas, 'DT_CONCESSAO_BOLSA'] = ""
```

```python
# Verificação da quantidade e das informações adicionadas
print("Valores únicos da coluna 'Data de concessão da bolsa': \n", df_2017['DT_CONCESSAO_BOLSA'].value_counts().head(5))
print()
print("Valores únicos da coluna 'Semestre de concessão da bolsa': \n", df_2017['SEMESTRE_CONCESSAO_BOLSA'].value_counts().head(5))
```
