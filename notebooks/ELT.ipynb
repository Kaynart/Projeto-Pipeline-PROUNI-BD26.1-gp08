```python
import pandas as pd
from sqlalchemy import create_engine
# --- CONFIGURAR A CONEXÃO COM O SEU POSTGRES ---
usuario_db = 'postgres'
senha_db = 'minhasenha'          # <- Coloque a sua senha aqui (com as aspas)
host_db = 'localhost'
porta_db = '5432'
nome_banco_db = 'prouni_elt'    # <- Se criou um banco específico (ex: prouni), mude aqui

DATABASE_URL = f'postgresql://{usuario_db}:{senha_db}@{host_db}:{porta_db}/{nome_banco_db}'
engine = create_engine(DATABASE_URL)
print("Conexão com o PostgreSQL criada com sucesso!")
```

```python
print("Iniciando processo de Extração e Carga (EL) do PROUNI...")

# --- MAPEAMENTO DOS SEUS ARQUIVOS DO PROUNI ---
arquivos_prouni = {
    'raw_prouni_2017': 'data/pda-prouni-2017.csv',
    'raw_prouni_2018': 'data/pda-prouni-2018.csv',
    'raw_prouni_2019': 'data/pda-prouni-2019.csv'
}

for nome_tabela, caminho_arquivo in arquivos_prouni.items():
    try:
        print("-" * 50)
        print(f"1. Extraindo: {caminho_arquivo}")
        
        # Lendo os dados brutos. Caso o arquivo use ponto e vírgula, adicionamos sep=';'
        # Adicionei também o encoding ISO-8859-1 (comum em dados do governo brasileiro)
        df_raw = pd.read_csv(
            caminho_arquivo, 
            sep=';',  
            encoding='ISO-8859-1',
            on_bad_lines='skip'
        )
        
        print(f"2. Carga: Enviando {len(df_raw)} linhas para a tabela {nome_tabela}...")
        
        # Salvando no Postgres na tabela correspondente
        df_raw.to_sql(
            nome_tabela,
            con=engine,
            if_exists='replace',
            index=False
        )
        print(f" Sucesso! Tabela {nome_tabela} carregada.")
        
    except Exception as e:
        print(f" Erro ao processar o arquivo {caminho_arquivo}: {e}")

print("\n Processo de Carga Bruta Concluído!")
```
